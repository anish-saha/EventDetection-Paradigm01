```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest 
```

```python
crypto = pd.read_csv('/Users/sabrinaa_le/Downloads/crypto-markets.csv')
btc = crypto.loc[crypto['name'] == 'Bitcoin']
btc = btc.drop(['slug', 'market', 'ranknow', 'volume', 'close_ratio', 'spread'], axis = 1)
btc
```

```python
btc.describe()
```

```python
btc.plot(x='date', y='close', figsize=(12,6))
plt.title('Bitcoin Prices over 5 Years')
plt.xlabel('Date')
plt.ylabel('Price of Bitcoin')
```

## Isolation Forest

```python
data = btc[['close']]
outliers_fraction=0.0315
scaler = StandardScaler()
np_scaled = scaler.fit_transform(data)
data = pd.DataFrame(np_scaled)

# train isolation forest
model =  IsolationForest(contamination=outliers_fraction)
model.fit(data) 
btc['anomaly2'] = pd.Series(model.predict(data))

# visualization
fig, ax = plt.subplots(figsize=(10,6))

#anomaly
a = btc.loc[btc['anomaly2'] == -1, ['date', 'close']]

ax.plot(btc['date'], btc['close'], color='blue', label = 'Normal')
ax.scatter(a['date'],a['close'], color='red', label = 'Anomaly')
plt.legend()
plt.title("Anamoly Detection Using Isolation Forest")
plt.xlabel('Date')
plt.ylabel('BTC Closing Price')
plt.show();
```

## Fitting an ARIMA Model

```python
#logging the clsing price
btc['log_close'] = np.log(btc['close'])
btc
```

```python
#time series data

from statsmodels.tsa.arima_model import ARIMA
#https://machinelearningmastery.com/arima-for-time-series-forecasting-with-python/

aic = 0
for i in range(0,4):
    for j in range(0,4):
        try:
            model = ARIMA(btc["log_close"], order=(i,0,j))
            model_fit = model.fit(disp=0)
            print("model converged for",i,j)
        except:
            model = ARIMA(btc["log_close"], order=(0,0,0))
            model_fit = model.fit(disp=0)
            print("model did not converge for",i,j)
            
        
        if model_fit.aic<aic:
            print("model_fit_aic",model_fit.aic)
            aic = model_fit.aic
            best_parameter_i = i
            best_parameter_j = j
        
        
        
model = ARIMA(btc["log_close"], order=(best_parameter_i,0,best_parameter_j))
model_fit = model.fit(disp=0)


print(model_fit.summary())
print(model_fit.aic)
```
