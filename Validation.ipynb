```python
placeholder = np.arange(1798) # ARIMA and IsolationForest have the same data for now.
df = pd.DataFrame({"ARIMA": placeholder, "IsolationForest": placeholder,
                   "KMeans": placeholder, "Naive_2SD": placeholder,
                   "Naive_Percent": placeholder})
df.head()
```

```python
name = 'isEvent_'
for file in os.listdir("./"):
    if file.endswith(".csv"):
        if name in file:
            print(file)
            data = pd.read_csv(file, index_col ='0')
            arr = data['0.1'].values
            col = file.replace(name, '').replace('.csv', '')
            df[col] = arr
df.describe()
```

```python
def similarity(df, col_1, col_2):
    #num_total = max(df[col_1].value_counts()[1], df[col_2].value_counts()[1])
    #num_incorrect = len(df.loc[df[col_1] != df[col_2]])
    #print(num_incorrect)
    #print(num_total)
    num_correct = len(df.loc[df[col_1] == df[col_2]])
    return num_correct / len(df)

result = np.array([])
for i in df.columns:
    temp = np.array([])
    for j in df.columns:
        temp = np.append(temp, similarity(df, i, j))
    result = np.append(result, temp)
        
result
```
